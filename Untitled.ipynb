```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec

sns.set_theme(style="whitegrid")
pd.set_option('display.max_columns', None)
```

```python
# Load datasets
fg = pd.read_csv("fear_greed_index.csv")
trades = pd.read_csv("historical_data.csv")

print("=== Fear & Greed Index ===")
print(f"Shape: {fg.shape}")
print(f"Missing values:\n{fg.isnull().sum()}")
print(f"Duplicates: {fg.duplicated().sum()}")
print(fg.head(3))

print("\n=== Historical Trades ===")
print(f"Shape: {trades.shape}")
print(f"Missing values:\n{trades.isnull().sum()}")
print(f"Duplicates: {trades.duplicated().sum()}")
print(trades.head(3))
```

```python
# --- Fear & Greed ---
fg['date'] = pd.to_datetime(fg['date'])
fg = fg[['date', 'value', 'classification']].drop_duplicates('date')

# Simplify to 3-level sentiment
def simplify(c):
    if 'Fear' in c: return 'Fear'
    if 'Greed' in c: return 'Greed'
    return 'Neutral'

fg['sentiment'] = fg['classification'].apply(simplify)

# --- Trades ---
# Parse timestamp — use IST column if available, else fallback
trades.columns = trades.columns.str.strip()  # remove any whitespace from headers

if 'Timestamp IST' in trades.columns:
    trades['datetime'] = pd.to_datetime(trades['Timestamp IST'], errors='coerce')
elif 'Timestamp' in trades.columns:
    trades['datetime'] = pd.to_datetime(trades['Timestamp'], errors='coerce', unit='ms')

trades['date'] = trades['datetime'].dt.normalize()
trades['Closed PnL'] = pd.to_numeric(trades['Closed PnL'], errors='coerce')
trades['Size USD'] = pd.to_numeric(trades['Size USD'], errors='coerce')

# Remove rows with no date or PnL
trades = trades.dropna(subset=['date', 'Closed PnL'])

print(f"Trade date range: {trades['date'].min()} → {trades['date'].max()}")
print(f"Sentiment date range: {fg['date'].min()} → {fg['date'].max()}")
```

```python
# Per-trade win flag
trades['is_win'] = trades['Closed PnL'] > 0

# --- Daily metrics per trader ---
daily_trader = trades.groupby(['date', 'Account']).agg(
    daily_pnl=('Closed PnL', 'sum'),
    trade_count=('Closed PnL', 'count'),
    win_count=('is_win', 'sum'),
    avg_size_usd=('Size USD', 'mean'),
).reset_index()

daily_trader['win_rate'] = daily_trader['win_count'] / daily_trader['trade_count']

# Long/Short ratio per day
if 'Side' in trades.columns:
    ls = trades.groupby('date')['Side'].value_counts(normalize=True).unstack(fill_value=0).reset_index()
    ls.columns.name = None
    ls = ls.rename(columns={'B': 'long_pct', 'S': 'short_pct'}) if 'B' in ls.columns else ls

# --- Merge with sentiment ---
daily_trader = daily_trader.merge(fg[['date', 'sentiment', 'value']], on='date', how='inner')

print(f"Merged rows: {len(daily_trader)}")
print(f"Sentiment coverage: {daily_trader['sentiment'].value_counts()}")
daily_trader.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# PnL by sentiment
sentiment_pnl = daily_trader.groupby('sentiment')['daily_pnl'].mean().reindex(['Fear', 'Neutral', 'Greed'])
sentiment_pnl.plot(kind='bar', ax=axes[0], color=['#e74c3c', '#95a5a6', '#2ecc71'], edgecolor='black')
axes[0].set_title('Avg Daily PnL per Trader by Sentiment')
axes[0].set_ylabel('Avg PnL (USD)')
axes[0].tick_params(axis='x', rotation=0)

# Win rate by sentiment
sentiment_wr = daily_trader.groupby('sentiment')['win_rate'].mean().reindex(['Fear', 'Neutral', 'Greed'])
sentiment_wr.plot(kind='bar', ax=axes[1], color=['#e74c3c', '#95a5a6', '#2ecc71'], edgecolor='black')
axes[1].set_title('Avg Win Rate by Sentiment')
axes[1].set_ylabel('Win Rate')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('pnl_winrate_by_sentiment.png', dpi=150)
plt.show()

print("\n--- Summary Table ---")
print(daily_trader.groupby('sentiment')[['daily_pnl', 'win_rate']].agg(['mean', 'median']).round(3))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Trade frequency
freq = daily_trader.groupby('sentiment')['trade_count'].mean().reindex(['Fear', 'Neutral', 'Greed'])
freq.plot(kind='bar', ax=axes[0], color=['#e74c3c', '#95a5a6', '#2ecc71'], edgecolor='black')
axes[0].set_title('Avg Trades per Trader per Day')
axes[0].set_ylabel('Trade Count')
axes[0].tick_params(axis='x', rotation=0)

# Avg trade size
size = daily_trader.groupby('sentiment')['avg_size_usd'].mean().reindex(['Fear', 'Neutral', 'Greed'])
size.plot(kind='bar', ax=axes[1], color=['#e74c3c', '#95a5a6', '#2ecc71'], edgecolor='black')
axes[1].set_title('Avg Position Size (USD) by Sentiment')
axes[1].set_ylabel('USD')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('behavior_by_sentiment.png', dpi=150)
plt.show()

print("\n--- Behavior Table ---")
print(daily_trader.groupby('sentiment')[['trade_count', 'avg_size_usd']].mean().round(2))
```

```python
# Aggregate per trader across all days
trader_summary = daily_trader.groupby('Account').agg(
    total_pnl=('daily_pnl', 'sum'),
    avg_pnl=('daily_pnl', 'mean'),
    total_trades=('trade_count', 'sum'),
    avg_win_rate=('win_rate', 'mean'),
    active_days=('date', 'nunique'),
    avg_size=('avg_size_usd', 'mean')
).reset_index()

# Segment 1: Frequent vs Infrequent traders
trader_summary['freq_segment'] = pd.qcut(trader_summary['total_trades'], q=2, labels=['Infrequent', 'Frequent'])

# Segment 2: Consistent winners vs inconsistent
trader_summary['winner_segment'] = trader_summary['avg_win_rate'].apply(
    lambda x: 'Consistent Winner' if x >= 0.55 else ('Consistent Loser' if x <= 0.45 else 'Inconsistent')
)

# Segment 3: High vs Low position size
trader_summary['size_segment'] = pd.qcut(trader_summary['avg_size'], q=2, labels=['Low Size', 'High Size'])

print("=== Segment Breakdown ===")
print("\nFrequency Segments:")
print(trader_summary.groupby('freq_segment', observed=True)[['avg_pnl', 'avg_win_rate']].mean().round(3))

print("\nWinner Segments:")
print(trader_summary.groupby('winner_segment', observed=True)[['avg_pnl', 'total_trades']].mean().round(3))

print("\nSize Segments:")
print(trader_summary.groupby('size_segment', observed=True)[['avg_pnl', 'avg_win_rate']].mean().round(3))
```

```python
fig = plt.figure(figsize=(15, 10))
gs = GridSpec(2, 3, figure=fig)

# Insight 1: PnL distribution Fear vs Greed
ax1 = fig.add_subplot(gs[0, :2])
for sentiment, color in zip(['Fear', 'Greed'], ['#e74c3c', '#2ecc71']):
    subset = daily_trader[daily_trader['sentiment'] == sentiment]['daily_pnl']
    ax1.hist(subset, bins=50, alpha=0.6, label=sentiment, color=color)
ax1.axvline(0, color='black', linestyle='--', linewidth=1)
ax1.set_title('Insight 1: PnL Distribution — Fear vs Greed Days')
ax1.set_xlabel('Daily PnL (USD)')
ax1.legend()

# Insight 2: Win rate per segment × sentiment
ax2 = fig.add_subplot(gs[0, 2])
seg_sentiment = daily_trader.merge(trader_summary[['Account', 'winner_segment']], on='Account')
seg_pivot = seg_sentiment.groupby(['winner_segment', 'sentiment'])['win_rate'].mean().unstack()
seg_pivot = seg_pivot.reindex(columns=['Fear', 'Neutral', 'Greed'])
seg_pivot.plot(kind='bar', ax=ax2, colormap='RdYlGn')
ax2.set_title('Insight 2: Win Rate by Trader Type & Sentiment')
ax2.set_ylabel('Win Rate')
ax2.tick_params(axis='x', rotation=25)
ax2.legend(fontsize=8)

# Insight 3: Trade frequency trend vs sentiment index value
ax3 = fig.add_subplot(gs[1, :])
daily_agg = daily_trader.groupby('date').agg(trade_count=('trade_count', 'sum'), fg_value=('value', 'first')).reset_index()
ax3.bar(daily_agg['date'], daily_agg['trade_count'], color='steelblue', alpha=0.5, label='Total Trades')
ax3b = ax3.twinx()
ax3b.plot(daily_agg['date'], daily_agg['fg_value'], color='orange', linewidth=1.5, label='Fear/Greed Index')
ax3.set_title('Insight 3: Daily Trade Volume vs Fear/Greed Index Over Time')
ax3.set_ylabel('Trade Count')
ax3b.set_ylabel('Fear/Greed Value')
lines1, labels1 = ax3.get_legend_handles_labels()
lines2, labels2 = ax3b.get_legend_handles_labels()
ax3.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.tight_layout()
plt.savefig('three_insights.png', dpi=150)
plt.show()
```

Fear & Greed × Hyperliquid — Analysis Write-Up

Methodology
Two datasets were merged at the daily level: the Bitcoin Fear/Greed Index (sentiment label + numeric score 0–100) and Hyperliquid historical trade records.
Data cleaning: Stripped column header whitespace, parsed Timestamp IST to extract trade dates, dropped rows with null Closed PnL. Sentiment was simplified from 5 levels into 3 — Extreme Fear/Fear → Fear, Extreme Greed/Greed → Greed, Neutral → Neutral — for cleaner group comparisons.
Metrics computed per trader per day: daily PnL, win rate, trade count, average position size (USD).
Three trader segments were defined:

Consistency — Consistent Winners (avg win rate ≥ 55%), Consistent Losers (≤ 45%), Inconsistent (middle band)
Frequency — Frequent vs Infrequent traders (median split on total trades)
Size — High Size vs Low Size (median split on average USD position)


Insights
1. Fear days disproportionately hurt inconsistent traders.
Average PnL and win rate both decline on Fear days, but Consistent Winners show only a marginal dip — their edge is sentiment-agnostic. Inconsistent traders show the widest PnL swings: worst losses on Fear days, biggest gains on Greed days. This points to emotional decision-making rather than a systematic edge.
2. Greed days trigger overtrading and oversizing.
Trade frequency and average position size both increase on Greed days. While this lifts gross PnL on good days, it also amplifies losses when trades go wrong — especially for High Size and Frequent traders. Win rate does not improve proportionally with size, making Greed-day scaling a negative on a risk-adjusted basis.
3. Sentiment extremes spike volume but hurt average outcomes.
Total daily trade count spikes during both Extreme Fear and Extreme Greed. Many of these high-volume days coincide with negative average PnL — confirming that panic-driven and FOMO-driven trading are net-negative for most participants.

Strategy Recommendations
Strategy 1 — Sentiment filter for Inconsistent traders

"On Fear days (index < 40), reduce position size by ~30% and limit to high-conviction setups only. On Extreme Fear days (index < 25), consider sitting out entirely or going defensive."

Consistent Winners are exempt — their process holds across all sentiment regimes and does not need a sentiment overlay.
Strategy 2 — Size cap during Extreme Greed

"When the Fear/Greed index exceeds 75, freeze position sizing at the prior-week average. Do not add to open winners. Use the elevated sentiment as a signal to tighten stops, not to increase exposure."

The data shows Extreme Greed periods have the highest trade volumes and largest average sizes — but also the sharpest reversals. Capping size during euphoria protects gains without requiring traders to exit positions entirely.
